```python
months = np.arange(1,12)
print(months)
```

```python
income = 100 + 3*months
print(income)
```

```python
plt.plot(months,income,'go')
```

```python
cost = 50 + 10*months
plt.plot(months,income,'go',months,cost,'r^')
```
