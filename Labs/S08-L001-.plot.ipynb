```python
import datetime
import pandas as pd
import matplotlib as mpl
%matplotlib inline

nasa = pd.read_csv("../Labs files//course-files/nasa_facebook_statuses.csv",
                    usecols=["status_published","num_reactions", "num_comments",
                             "num_shares","num_likes"])

nasa["status_published"] = pd.to_datetime(nasa["status_published"])

filter =(nasa['status_published'] >= '2016-5-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]

nasa['day'] = nasa.apply(lambda row: row["status_published"].day,axis=1)

nasa_by_day = nasa.groupby(by='day').sum()
nasa_by_day.head()
```

```python
nasa_by_day.plot()
```

```python
nasa_by_day.plot(y='num_shares')
```
