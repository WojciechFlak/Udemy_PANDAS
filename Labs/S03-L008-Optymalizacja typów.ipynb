```python
fuel = pd.read_csv('../Labs files/course-files/fuel.csv', low_memory=False,
                  usecols=['Year','Make','Model','Class',
                           'Fuel Type','Combined MPG (FT1)'],
                  dtype={'Make': 'category', "Fuel Type": 'category', 'Year': 'int16'})
fuel.head()
```

```python
fuel.info(memory_usage='deep')
```

```python
fuel['Make'].value_counts()
```

```python
fuel['Combined MPG (FT1)'] = fuel['Combined MPG (FT1)'].astype('float32')
fuel.info(memory_usage='deep')
```
