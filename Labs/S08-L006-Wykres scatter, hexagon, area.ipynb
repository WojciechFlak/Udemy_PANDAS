```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline

marathon = pd.read_csv("../Labs files/course-files/marathon_results_2017.csv", 
                       usecols=["Age","M/F","Country","40K"])
marathon.dropna(how='any')
marathon["TimeSeconds"] = marathon["40K"].apply(lambda x: \
                        pd.to_timedelta(x,errors='coerce').total_seconds())
marathon.head(5)
```

```python
marathon.plot(kind='scatter',x='Age',y='TimeSeconds')

```

```python
marathon_M = marathon[marathon['M/F']=='M']
```

```python
marathon_F = marathon[marathon['M/F']=='F']
```

```python
ax = marathon_M.plot(kind='scatter',x='Age',y='TimeSeconds',label='M')
marathon_F.plot.scatter(x="Age",y="TimeSeconds",color="Pink",ax=ax,label='F',alpha=0.2)
```

```python
marathon_F.plot.hexbin(x='Age',y='TimeSeconds',gridsize=10)
```

```python
marathon_M.plot.hexbin(x='Age',y='TimeSeconds', gridsize=15)
```

```python
area = marathon_M[['Age','TimeSeconds']]
```

```python
area.head()
```

```python
area_F = marathon_F[['Age','TimeSeconds']]
area_F.head()
```

```python
area = area.merge(area_F, left_on='Age', right_on='Age',
                  suffixes=['_man','_woman'],how='outer')
area.fillna(0)
area.set_index('Age',inplace=True)
```
