```python
import pandas as pd

f_2015 = pd.read_csv('../Labs files/course-files/Fortune_500_2015.csv',
                    usecols=['rank','company'])
f_2016 = pd.read_csv('../Labs files/course-files/Fortune_500_2016.csv',
                    usecols=['rank','company'])
f_2015.head()
```

```python
companies_15_16 = f_2015.merge(f_2016,how='inner',indicator=True, on='company')
companies_15_16.head()
```

```python
companies_15_16 = f_2015.merge(f_2016,how='inner',indicator=True, on='company',
                              suffixes=['_2015','_2016'])
companies_15_16.head()
```
