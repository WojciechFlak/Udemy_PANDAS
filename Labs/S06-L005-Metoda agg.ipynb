```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time

df = pd.read_csv('../Labs files/course-files/marathon_results_2016.csv',
                   usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],
                index_col='Bib')
df.head()
```

```python
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
sex_age = df.groupby(['M/F','Age'])
```

```python
sex_age.agg({'TotalSeconds': 'mean', 'HalfSeconds': 'mean'})
```

```python
sex_age.agg({'TotalSeconds': 'sum', 'HalfSeconds': 'sum'}).head()
```

```python
sex_age.agg({'TotalSeconds':    ['sum','mean'], 
             'HalfSeconds':     ['sum','mean']}).head()
```

```python
functions = ['mean','sum','count']
```

```python
sex_age.agg({'TotalSeconds':    functions, 
             'HalfSeconds':     functions}).head()
```
