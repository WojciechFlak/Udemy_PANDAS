```python
import pandas as pd
import numpy as np

fuel = pd.read_csv('../Labs files/course-files/fuel.csv',
                   usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'])
fuel.head()
```

```python
fuel.drop_duplicates(['Year','Make','Model'],inplace=True)
fuel = fuel.dropna(how='any')
fuel.set_index(['Year','Make','Model'],inplace=True)
fuel.head()
```
